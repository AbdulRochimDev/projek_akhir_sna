```python
import pandas as pd

# Read the Excel file into a dataframe
dataframe = pd.read_excel('/content/projek_akhir_sna/twomodegame.xlsx')
```

```python
import networkx as nx

graph2 = nx.Graph()

for index, row in dataframe.iterrows():
    siswax2 = row['Nama']
    sisway2 = row[1:]

    graph2.add_node(siswax2)

    for i, siswa2 in enumerate(sisway2):
        if siswa2>= 1:
            siswa_name = dataframe.columns[i + 1]
            graph2.add_node(siswa_name)
            graph2.add_edge(siswax2, siswa_name,weight=siswa2)

print(f"Number of nodes: {graph2.number_of_nodes()}")
print(f"Number of edges: {graph2.number_of_edges()}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a dictionary of edge labels with weights
edge_labels = {(u, v): weight for u, v, weight in graph2.edges.data('weight')}

plt.figure(figsize=(12, 8))  # Adjust the dimensions as desired
pos = nx.kamada_kawai_layout(graph2)

nx.draw_networkx_nodes(graph2, pos, node_color='lightblue', node_size=300)
nx.draw_networkx_edges(graph2, pos, edge_color='gray', width=1, alpha=0.5)
nx.draw_networkx_labels(graph2, pos, font_size=10)
nx.draw_networkx_edge_labels(graph2, pos, edge_labels=edge_labels, font_size=8)

plt.axis('off')
plt.tight_layout()

plt.show()
```

Lakukan Louvain/MCL Clustering dan laporkan list cluster-clusternya.

```python
from community import community_louvain
import matplotlib.cm as cm
import matplotlib.pyplot as plt
partition = community_louvain.best_partition(graph2)
pos = nx.spring_layout(graph2)
#cmap = cm.get_cmap('viridis', max(partition.values())+1)
cmap = cm.get_cmap('viridis', max(partition.values()) + 1)
nx.draw_networkx_nodes(graph2, pos, partition.keys(), node_size=40, cmap=cmap, node_color=list(partition.values()))
nx.draw_networkx_edges(graph2, pos, alpha=0.5)
plt.show()
```

```python
communities ={}
for node, community in partition.items():
  if community not in communities:
    communities[community]=[]
  communities[community].append(node)
for community, nodes in communities.items():
  print(f"community {community}:{nodes}")
```

# Nomer 5

```python
# Calculate the degree centrality of the nodes
deg_centrality = nx.degree_centrality(graph2)


# Sort degree centrality in descending order
sorted_deg_centrality = sorted(deg_centrality.items(), key=lambda x: x[1], reverse=True)

print("Degree Centrality (sorted)")
for node, value in sorted_deg_centrality:
    print(f"Node {node}: {value}")
print("")
```

```python
max_val = max(deg_centrality.values())
keys_val = [key for key, value in deg_centrality.items() if value == max_val]
print("Maximum Degree Centrality Siswa:")
print(f"{keys_val}: {max_val}\n")

min_val = min(deg_centrality.values())
keys_val = [key for key, value in deg_centrality.items() if value == min_val]
print("Minimum Degree Centrality Siswa:")
print(f"{keys_val}: {min_val}")
```

# Nomer 8
Temukan apakah Graph ini memenuhi paradigma Friendship Paradox

```python
import networkx as nx

# Assuming you already have the graph 'graph'

# Calculate the average degree of the nodes
degree_nodes = dict(graph2.degree())
avg_degree_nodes = sum(degree_nodes.values()) / len(degree_nodes)

# Calculate the average degree of the neighbors
avg_degree_neighbors = 0
for node in graph2.nodes():
    neighbors = graph2.neighbors(node)
    avg_degree_neighbors += sum(degree_nodes[n] for n in neighbors)
avg_degree_neighbors /= len(graph2)

# Check if the friendship paradox holds
if avg_degree_nodes < avg_degree_neighbors:
    print("Friendship Paradox holds")
else:
    print("Friendship Paradox does not hold")
```

```python
dataframe = pd.read_excel('/content/projek_akhir_sna/gameprocess.xlsx')
```

```python
all_nama = []

for i in dataframe["Nama"]:
  if i not in all_nama:
    all_nama.append(i)
```

```python
all_game = []

for index, row in dataframe.iterrows():
  if row["Game"] not in all_game:
    all_game.append(row["Game"])
```

```python
all_game[5]="Kosong"
```

```python
graph2 = nx.Graph()
nama = all_nama
game = all_game

graph2.add_nodes_from(nama, group="nama")
graph2.add_nodes_from(game, group="game")

for index, row in dataframe.iterrows():
  if(row["Nama"] in nama and row["Game"] in game):
    graph2.add_edge(row["Nama"], row["Game"])
```

```python
import random
pos = nx.kamada_kawai_layout(graph2)
colors = {"nama": "blue", "game": "green"}

left_nodes = [node for node in graph2.nodes if graph2.nodes[node]["group"] == "nama"]
right_nodes = [node for node in graph2.nodes if graph2.nodes[node]["group"] == "game"]


y_step_left = 1.0 / (len(left_nodes) + 1)

pos_left = {node: (-2, (i + 1) * y_step_left) for i, node in enumerate(left_nodes)}

y_range = 0.6
y_min = min(pos[node][1] for node in right_nodes)
y_max = max(pos[node][1] for node in right_nodes)

pos_right = {}

for node in right_nodes:
    y = pos[node][1]
    y_scaled = (y - y_min) / (y_max - y_min)
    y_shifted = y_scaled * y_range + (1 - y_range) / 5
    pos_right[node] = (2, y_shifted)


pos.update(pos_left)
pos.update(pos_right)


colors = {"nama": "blue", "game": "green"}
node_colors = [colors[graph2.nodes[node]["group"]] for node in graph2.nodes]
nx.draw_networkx(graph2, pos, with_labels=True, font_size=10, node_color=node_colors)


plt.show()
```
